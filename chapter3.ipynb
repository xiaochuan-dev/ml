# 神经网络

通过模型不断学习逼近现实中的函数。

在单变量输入和单变量输出中，每个神经元有

$$

f(w_ix+b_i)

$$

如果输入是多值的

```python
# 输入为单变量，输出为单变量
# 这里用二次函数做逼近

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 2000).reshape(-1, 1)
y = 2*x**2 + 3*x + 1 + np.random.normal(0, 0.1, x.shape)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),     # 只有一层隐藏层，总共64个隐藏单元
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(x, y, epochs=100, batch_size=32, verbose=0)

x_test = np.array([[-2], [0], [2]])
y_pred = model.predict(x_test)

plt.scatter(x, y, label='y', s=2)
plt.scatter(x_test, y_pred, label='y_pred', s=2)
plt.legend()
plt.show()
```

```python
# 这里测试e^x 这个函数

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200).reshape(-1, 1)
y = tf.exp(x) + np.random.normal(0, 0.1, x.shape)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(x, y, epochs=100, batch_size=32, verbose=0)

x_test = np.array([[-2], [0], [2]])
y_pred = model.predict(x_test)

plt.scatter(x, y, label='y', s=2)
plt.scatter(x_test, y_pred, label='y_pred', s=2)
plt.legend()
plt.show()
```

```python
# 这里测试e^x 这个函数
# 使用深度神经网络

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200).reshape(-1, 1)
y = tf.exp(x) + np.random.normal(0, 0.1, x.shape)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(x, y, epochs=100, batch_size=32, verbose=0)

x_test = np.array([[-2], [0], [2]])
y_pred = model.predict(x_test)

plt.scatter(x, y, label='y', s=2)
plt.scatter(x_test, y_pred, label='y_pred', s=2)
plt.legend()
plt.show()
```
